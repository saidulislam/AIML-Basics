```python
# Import Pandas for data processing
import pandas as pd
```

```python
# Read the CSV file
dataset = pd.read_csv('StudentsMLR.csv')
df = dataset.copy()
df
```

```python
# Split into X (Independent) and Y (predicted)
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

```python
X
```

```python
Y
```

```python
# Split the rows into training and test set
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =     \
train_test_split (X, Y, test_size = 0.2, random_state=42)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression

std_reg = LinearRegression()
```

```python
# training on the Data
std_reg.fit(x_train, y_train)
```

```python
# predict the results of test data
y_predict = std_reg.predict(x_test)
```

```python
# Look at the first few predictions
# Comparing predictions with actual results

print("Predicted:", y_predict[:5])
print("Actual   :", y_test[:5].values)
```

```python
# Get the R-Squared 
mlr_score = std_reg.score(x_test, y_test)
mlr_score
```

```python
# Mean Absolute Error (MAE)
# absolute difference between predicted and actual values
# Basically, it's telling you... on average, your predictions are off by X points
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_predict)
print("MAE:", mae)
```

```python
# Root Mean Squared Error (RMSE)
# tells you... your predictions are typically within +/- X points.
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_test, y_predict))
print("RMSE:", rmse)
```

```python
# "accuracy like" metric
accuracy = 100 - (mae / y_test.mean() * 100)
print("Approx Accuracy (%):", accuracy)
```

```python
# Coefficient and Intercept
mlr_coefficient_0 = std_reg.coef_[0]
mlr_coefficient_1 = std_reg.coef_[1]
mlr_intercept = std_reg.intercept_

print(mlr_coefficient_0)
print(mlr_coefficient_1)
print(mlr_intercept)
```
